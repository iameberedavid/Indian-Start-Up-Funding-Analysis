# Checking for custom quality
The cells of the "Amounts ($)" colummn of the dataset has amounts in rupees, amounts in dollars, and dash symbols. The rupee amounts will be converted to dollars since the amounts in other datasets are also in dollars. The average exchange rate of rupees to dollars in 2018 will be used to ensure accurate conversion. The dash symbols will be handled afterwards.

```python
# Before converting, let's check the datatypes of the values in the "Amount($)" column

df1["Amount($)"].apply(type).value_counts()
```

The rupee amounts alone need to be converted and the presence of the rupee symbol in the amounts suggest that the datatype would be string. There is need to change the datatypes to float to make the conversion possible. We will first remove all the non-numerical values then applying .astype(float). The non-numerical values are the rupee symbol and the comma symbol. While removing them, we will use the lambda function to ensure that NaN values are not returned.

```python
# Define the average exchange rate in 2018
exchange_rate = 0.0146  # That is, 1 Rupee = 0.0146 Dollar

# Select the rows with rupee amounts
rupee_rows = df1["Amount($)"].str.contains('₹') # Note that only the rupee amounts should be converted 

# Remove the non-numerical values and change the datatype to float
df1.loc[rupee_rows, "Amount($)"] = df1.loc[rupee_rows, "Amount($)"].apply(lambda x: x.replace('₹', '').replace(',', '')
                                if isinstance(x, str) else x).astype(float)

# Convert the rupee amounts to dollars
df1.loc[rupee_rows, "Amount($)"] = df1.loc[rupee_rows, "Amount($)"] * exchange_rate

df1
```

# Handling the dash symbols

```python
# Counting the number of cells containing the dash symbol

(df1["Amount($)"] == '—').sum()
```

The dash symbol in the amount column of 148 startups could mean that the start-ups received no funding, or that the actual data were lost during the data collection process. Since the dataset originally did not contain missing values, assuming that some rows have dash symbols because the actual data was lost during the data collection process will be unnecessary. It is safer to assume that the dash sumbol represent zero fuding. For this reason, the dash symbols will be replaced with zero (0).

```python
# Replacing the dash symbol with zero

df1['Amount($)'] = df1['Amount($)'].replace('—', '0', regex=True)
df1
```

```python
# Removing possible dollar and commma symbols in the entire Amount($) column

df1['Amount($)'] = df1['Amount($)'].replace(['\$', ','], '', regex=True)
df1
```

```python
# Changing the datatype of the Amount($) column back to integer

df1["Amount($)"] = df1["Amount($)"].astype(np.int64)

df1.head()
```

```python
# Rearranging the columns in the dataset to fit the format in the second, third and fourth datasets

df1 = df1.loc[:, ["Company/Brand", "HeadQuarter", "Sector", "What it does", "Amount($)", "Stage"]]
df1.head()
```

In the first dataset, the "Headquarter" column contains the names of the cities, states and country. But the other datasets have only  the names of cities in the "Headquarter" column. In order to ensure that the datasets maintain the same format, the names of states and country will be removed from the "Headquarter" column of the first dataset, leaving only the names of cities.

```python
# Split the column values by comma and select only the first word (names of cities)

df1["HeadQuarter"] = df1["HeadQuarter"].str.split(',').str[0]
df1
```
