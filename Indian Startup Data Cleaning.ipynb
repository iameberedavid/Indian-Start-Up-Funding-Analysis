```python
import pandas as pd

df1 = pd.read_csv('startup_funding2018.csv')
df2 = pd.read_csv('startup_funding2019.csv')
df3 = pd.read_csv('startup_funding2020.csv')
df4 = pd.read_csv('startup_funding2021.csv')
```

```python


column_dict = {
    'Company Name': 'Company/Brand',
    'Industry': 'Sector',
    'Round/Series': 'Stage',
    'Location': 'HeadQuarter',
    'About Company': 'What it does',
    'Amount': 'Amount($)'
}

df1 = df1.rename(columns=column_dict)

```

```python
df1.head()
```

# Converte all the currency into a single currency

```python
import re
exchange_rate = 0.01437 #Rupee <> $ Exchange Rate as at 2018 EOY

df1['Amount($)'] = df1['Amount($)'].replace('[₹,]', '', regex=True)
df1['Amount($)'] = df1['Amount($)'].replace('[$,]', '', regex=True)
df1['Amount($)'] = df1['Amount($)'].replace('[—,]', '0', regex=True)


df1['Amount($)'] = df1['Amount($)'].astype(float)

df1['Amount($)'] = df1['Amount($)'] * exchange_rate

df1['Amount($)'] = '$' + df1['Amount($)'].astype(str)



df1.head()
```

# Merge the data into a single dataset

```python

merged_df = pd.concat([df1, df2, df3, df4], ignore_index=True)
merged_df
```

# Drop the rows with null rows

```python
merged_df = merged_df.dropna()
merged_df.head()
```

# Drop the unnamed column

```python
merged_df = merged_df.drop('Unnamed: 9', axis=1)
merged_df.head()
```

```python
merged_df['Amount($)'] = merged_df['Amount($)'].replace('[Undisclosed,]', '', regex=True)

merged_df.head()
```

# Create a cleaned csv file

```python
merged_df.to_csv("merged_cleaned_file.csv")
```
